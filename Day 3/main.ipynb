```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(1000 * (1 + 0.10 * 5))          # Simple interest
print(1000 * (1 + 0.10/1) ** (1*5))   # Compound interest
print(1000 * np.exp(0.10 * 5))        # Continuous compounding
```

```python
# Constants
P = 1000
r = 0.10
t = np.linspace(0, 10, 100)  # 10 years

# Formulas
A_simple = P * (1 + r * t)
A_compound = P * (1 + r/1)**(1 * t)
A_continuous = P * np.exp(r * t)

# Plot
plt.figure(figsize=(8,5))
plt.plot(t, A_simple, label="Simple Interest", linestyle="--")
plt.plot(t, A_compound, label="Compound Interest")
plt.plot(t, A_continuous, label="Continuous Compounding", linestyle=":")
plt.title("Growth of $1000 Over Time at 10% Interest")
plt.xlabel("Time (years)")
plt.ylabel("Amount (â‚¹)")
plt.legend()
plt.grid(True)
plt.show()
```
