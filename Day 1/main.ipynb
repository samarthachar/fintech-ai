```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,100)# Generates 100 evenly spaced numbers from and including -10 and 100
y = 2*x + 5# Sets the equation as y = 2x + 5


plt.plot(x,y)# Function: f(x) = 2x + 5
plt.title("y = 2x + 5")

# Frames the graph
plt.xlim(0, 20)
plt.ylim(0, 20)

# Plotting multiple lines
plt.plot(x,x**2)
plt.plot(x,(3* x ** 2 - 2 * x - 1))

plt.show()
```
