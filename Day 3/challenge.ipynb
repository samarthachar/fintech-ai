```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
P = 1000.0

BankA_r = 0.098
BankB_r = 0.096
BankC_r = 0.094

t = 5

BankA_result = P * ( 1 + BankA_r) ** t
BankB_result = P * ( 1 + BankB_r/12) ** (12 * t)
BankC_result = P * np.exp(BankC_r * t)

print(f"Bank A after {t} years: {BankA_result}")
print(f"Bank B after {t} years: {BankB_result}")
print(f"Bank C after {t} years: {BankC_result}")

results = {
    "Bank A": BankA_result,
    "Bank B": BankB_result,
    "Bank C": BankC_result
}



difference = results[max(results, key=results.get)] - results[min(results, key=results.get)]
highest_bank = max(results, key=results.get)

print(f"{highest_bank} had the most by a difference of {difference}")

```

```python
print(f"EAR of Bank A: {(1 + BankA_r/1) ** 1 - 1}")
print(f"EAR of Bank B: {(1 + BankB_r/12) ** 12 - 1}")
print(f"EAR of Bank C: {np.exp(BankC_r) - 1}")
```

```python
# Redifine variables
t = np.linspace(0, 100, 200)

BankA_result = P * ( 1 + BankA_r) ** t
BankB_result = P * ( 1 + BankB_r/12) ** (12 * t)
BankC_result = P * np.exp(BankC_r * t)

# --- Plot ---
plt.figure(figsize=(9,6))
plt.plot(t, BankA_result, label="Bank A (Annual 9.8%)", linestyle="--", linewidth=2)
plt.plot(t, BankB_result, label="Bank B (Monthly 9.6%)", linewidth=2)
plt.plot(t, BankC_result, label="Bank C (Continuous 9.4%)", linestyle=":", linewidth=2)

# Titles and labels
plt.title("Growth of $1000 Over 100 Years at Different Interest Types", fontsize=14)
plt.xlabel("Time (years)", fontsize=12)
plt.ylabel("Amount ($)", fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)

plt.show()
```
