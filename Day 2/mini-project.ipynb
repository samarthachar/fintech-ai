```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
years = np.arange(0, 11) # Gives array from 0 to 10
principal = 1000 #principal; ₹1000
rate = 0.08 # 5%
```

```python
simple_interest = principal * (1 + rate * years)
compound_interest = principal * (1 + rate)**years
```

```python
plt.figure(figsize=(8,4))
plt.plot(years, simple_interest, marker='o', label='Simple Interest')
plt.plot(years, compound_interest, marker='o', label='Compound Interest')
plt.title('Simple vs Compound Interest (₹1000 at 5%)')
plt.xlabel('Years')
plt.ylabel('Amount (₹)')
plt.legend()
plt.grid(True)
plt.show()
```
