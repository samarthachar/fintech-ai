```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(12)
returns = np.random.normal(0.0005, 0.01, 252)
price = (1 + returns).cumprod() * 100

MA20 = pd.Series(price).rolling(20).mean()

plt.figure(figsize=(10,5))
plt.plot(price, label="Price", color="blue")
plt.plot(MA20, label="20-Day MA", color="orange")

plt.title("Price vs 20-Day Moving Average")
plt.xlabel("Days")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```
