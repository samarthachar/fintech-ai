```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create x values
x = np.linspace(-10, 10, 400)

# define functions
y_linear = 2 * x + 3        # linear
y_quadratic = x**2          # nonlinear (parabola)
y_sin = np.sin(x)           # periodic / nonlinear
```

```python
# plot them together

plt.figure(figsize=(9,5))

plt.plot(x, y_linear, label='y = 2x + 3')
plt.plot(x, y_quadratic, label='y = x')
plt.plot(x, y_sin, label='y = sin(x)')

plt.legend()
plt.title('Linear vs Nonlinear Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.ylim(-10, 100)
plt.show()
```
