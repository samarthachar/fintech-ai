```python
import numpy as np
import matplotlib.pyplot as plt

r = 0.10  # 10% nominal annual rate
n_values = np.array([1, 2, 4, 12, 52, 365, 10000])  # compounding frequencies

EAR_values = (1 + r / n_values)**n_values - 1
EAR_continuous = np.exp(r) - 1

plt.figure(figsize=(9,6))
plt.plot(n_values, EAR_values * 100, "o-", label="Discrete Compounding")
plt.axhline(EAR_continuous * 100, color="red", linestyle="--", label="Continuous Limit")

plt.xscale("log")  # makes frequency clearer
plt.xlabel("Compounding Frequency (n per year)", fontsize=12)
plt.ylabel("Effective Annual Rate (%)", fontsize=12)
plt.title("EAR vs Compounding Frequency", fontsize=14)
plt.grid(True, which="both", linestyle=":", alpha=0.5)
plt.legend()
plt.show()
```

```python
P = 1000
r = 0.10
t = np.linspace(0, 10, 100)

# Annual compounding
A_discrete = P * (1 + r)**t

# Continuous compounding
A_continuous = P * np.exp(r * t)

plt.figure(figsize=(9,6))
plt.plot(t, A_discrete, label="Annual Compounding", linewidth=2)
plt.plot(t, A_continuous, label="Continuous Compounding", linestyle="--", linewidth=2)

plt.title("Discrete vs Continuous Compounding Growth (10 Years)", fontsize=14)
plt.xlabel("Time (years)", fontsize=12)
plt.ylabel("Amount ()", fontsize=12)
plt.legend()
plt.grid(True, alpha=0.4)
plt.show()
```

```python
t = np.linspace(0, 10, 100)
learning_rate = 0.3  # try changing to 0.1 or 0.7 later

performance = 1 - np.exp(-learning_rate * t)

plt.figure(figsize=(9,6))
plt.plot(t, performance, linewidth=2, color='royalblue')
plt.title("AI Learning Curve (Continuous Improvement)", fontsize=14)
plt.xlabel("Time (training steps)", fontsize=12)
plt.ylabel("Model Performance (01)", fontsize=12)
plt.grid(True, alpha=0.4)
plt.show()
```

```python
t = np.linspace(0, 10, 100)
r = 0.3
performance = 1 - np.exp(-r * t)
money_growth = np.exp(r * t) / np.exp(r * 10)

plt.figure(figsize=(9,6))
plt.plot(t, performance, label="AI Learning", linewidth=2)
plt.plot(t, money_growth, label="Money Growth", linestyle="--", linewidth=2)
plt.title("AI Learning vs Continuous Compounding", fontsize=14)
plt.xlabel("Time", fontsize=12)
plt.ylabel("Normalised Value", fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
